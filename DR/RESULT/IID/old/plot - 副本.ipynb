```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
os.listdir()
```

```python
val = []
train = []
with open('dr456s.out','r') as f: #efficientnetb5+transfer+softmax last layer
    for line in f:
#         if line[:5] == '33/33':
        if line[:3] == '8/8':
            val.append(float(line.split(' ')[-1]))
            train.append(float(line.split(' ')[-7]))
epo = [i+1 for i in range(len(val))]
plt.plot(epo,val,label='val_acc')
plt.plot(epo,train,label='train_acc')
plt.legend()
```
