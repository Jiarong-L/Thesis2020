```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
val = []
train = []
val_loss = []
train_loss = []
with open('cen.txt','r') as f: #VGG16- weights=None
    for line in f:
        if line[:5] == '80/80':
            a=line
            val.append(float(line.split(' ')[-1]))
            train.append(float(line.split(' ')[-7]))
            val_loss.append(float(line.split(' ')[-4]))
            train_loss.append(float(line.split(' ')[-10]))
epo = [i+1 for i in range(len(val))]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=False, figsize=(12, 4))


ax[0].plot(epo,val,label='val_acc')
ax[0].plot(epo,train,label='train_acc')
ax[0].legend()
ax[1].plot(epo,val_loss,label='val_loss')
ax[1].plot(epo,train_loss,label='train_loss')
ax[1].legend()
ax[0].grid(axis="y",ls='-.')
ax[1].grid(axis="y",ls='-.')
```

```python
acc = []
loss = []
with open('fed.txt','r') as f: #VGG16- weights=None
    for line in f:
        if line[:7] == '200/200':
            a=line
            acc.append(float(line.split(' ')[-1][:-1]))
            loss.append(float(line.split(' ')[-4]))
epo_fed = [i+1 for i in range(len(acc))]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=False, figsize=(12, 4))


ax[0].plot(epo[:500],val[:500],label='Central_acc')
ax[0].plot(epo_fed[:500], acc[:500],label='Federated_acc')
# ax[0].set_ylim([0.62,0.69])
ax[0].legend()
ax[1].plot(epo[:500],val_loss[:500],label='Central_loss')
ax[1].plot(epo_fed[:500], loss[:500],label='Federated_loss')
ax[1].legend()
ax[0].grid(axis="y",ls='-.')
ax[1].grid(axis="y",ls='-.')
```
