```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=False, figsize=(12, 4))

val = []
train = []
val_loss = []
train_loss = []
with open('xxxxxx.txt','r') as f: #VGG16- weights=None
# with open('old/aug_testingset/dr224.out','r') as f: #VGG16- weights=None
    for line in f:
        if line[:7] == '200/200':
            a=line
            val.append(float(line.split(' ')[-1]))
            train.append(float(line.split(' ')[-7]))
            val_loss.append(float(line.split(' ')[-4]))
            train_loss.append(float(line.split(' ')[-10]))
epo = [i+1 for i in range(300)]
ax[0].plot(epo,val[:300],label='val_acc')
ax[0].plot(epo,train[:300],label='train_acc')
ax[0].legend()
ax[1].plot(epo,val_loss[:300],label='val_loss')
ax[1].plot(epo,train_loss[:300],label='train_loss')
ax[1].legend()
ax[0].grid(axis="y",ls='-.')
ax[1].grid(axis="y",ls='-.')
```
