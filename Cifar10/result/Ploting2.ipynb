```python
import numpy as np
import os
import matplotlib.pyplot as plt
import glob
import statistics
%matplotlib inline
```

```python
def es_point(s,p):
    max_index=np.where(s[2]==np.amax(s[2]))
    max_index=max_index[0]+p
    return max_index[0]
```

```python
def windowed(s,window_size):
    s1=[]
    sv=[]
    for i in range(len(s)-window_size):
        d=s[i:i+window_size]
        dm=sum(d)/window_size
        s1.append(dm)
        dv=statistics.variance(d)
        sv.append(dv)
    return s1,sv
```
